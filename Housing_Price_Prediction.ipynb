```python
df.drop('street',axis='columns', inplace=True)
df.drop('city',axis='columns', inplace=True)
df.drop('statezip',axis='columns', inplace=True)
df.drop('country',axis='columns', inplace=True)
```

```python
from sklearn.model_selection import train_test_split

x = df[['bedrooms','bathrooms','sqft_living','sqft_lot','floors','condition','view','sqft_basement']]
y = df['price']

X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
Y_test
```

```python
Y_train
```

```python
from sklearn.linear_model import LinearRegression

m = LinearRegression()
m.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
y_prediction = m.predict(X_test)
```

```python
a = mean_squared_error(Y_test,y_prediction)
b = r2_score(Y_test,y_prediction)

print(a)
print(b)
```

```python
plt.scatter(Y_test,y_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

```python
plt.scatter(Y_test,y_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
```

```python
residuals = Y_test-y_prediction
plt.scatter(Y_test,residuals)
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
