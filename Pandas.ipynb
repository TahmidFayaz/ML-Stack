```python
#Applying Functions To DataFrame

stuff = {
    'Corporation':['Apple', 'Google', 'Meta', 'Apple', 'Google','Meta'],
    'Employees':['John', 'April', 'Wes', 'Beth', 'Justin', 'Steph'],
    'Salary':[200, 220, 190, 130, 120, 150]}
# Create Dataframe
my_df = pd.DataFrame(stuff)
my_df
```

```python
#Create Function
def times20(x):
    return x*20
```

```python
#Use Apply Function
my_df["Salary"].apply(times20)
```

```python
my_df["Salary"] = my_df["Salary"].apply(times20)
my_df
```

```python
#Linecharts

import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(25,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Line plot
my_df.plot(kind="line")
```

```python
# line plot absolute
my_df.abs().plot(kind="line")
```

```python
# Line single
my_df['Mon'].abs().plot(kind="line")
```

```python
# Line More than one
my_df[['Mon', 'Tues']].abs().plot(kind="line")
```

```python
# Size
my_df[['Mon', 'Tues']].abs().plot(kind="line", figsize=(10,5))
```

```python
# Title
my_df[['Mon', 'Tues']].abs().plot(kind="line", title="My Awesome Line Chart")
```
