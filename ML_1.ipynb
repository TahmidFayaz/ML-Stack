Data Seperation


```python
y= DF["logS"]
y
```

```python
x=DF.drop("logS",axis=1)
x
```

Dataset Seperation


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.3,random_state=100)

x_train
```

```python
x_test
```

Linear Regression


```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
Y_LR_Prediction_Train=LR.predict(x_train)
Y_LR_Prediction_Test=LR.predict(x_test)
```

```python
print(Y_LR_Prediction_Train)
```

```python
print(Y_LR_Prediction_Test)
```

Model Performance


```python
y_train
```

```python
Y_LR_Prediction_Train
```

```python

from sklearn.metrics import mean_squared_error,r2_score


LR_Train_MSE=mean_squared_error(y_train,Y_LR_Prediction_Train)
LR_Train_R2=r2_score(y_train,Y_LR_Prediction_Train)

LR_Test_MSE=mean_squared_error(y_test,Y_LR_Prediction_Test)
LR_Test_R2=r2_score(y_test,Y_LR_Prediction_Test)
```

```python
print("LR_Train_MSE",LR_Train_MSE)
print("LR_Train_R2",LR_Train_R2)
print("LR_Test_MSE",LR_Test_MSE)
print("LR_Test_R2",LR_Test_R2)
```
