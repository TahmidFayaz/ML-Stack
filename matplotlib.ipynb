```python
plt.title("First graph")
x=[32,45,67]
y=[21,24,2300]

plt.plot(x,y)
plt.show()
#title
```

```python
#x-label,y-label
plt.title("First graph",fontdict={'fontsize':20})
x=[32,45,67]
y=[21,24,2300]
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(x,y)
plt.show()

#xsticks,ysticks ---> sizing of the graph
```

```python
plt.plot([1,2,3],[4,5,6],label="2X",color='#301934',linewidth=8)
plt.legend()
```

```python
plt.plot([1,2,3],[4,5,6],label="2X",color="violet",marker=".",markersize=20,markeredgecolor="blue",linestyle='--')
plt.legend()
```

```python
#Shortcut
x=[2,4,6]
y=[8,10,12]

plt.plot(x,y,'b.--',label="2x")
#color,marker,lineswidth
```

```python
m=np.arange(0,4,0.5)
x=[2,4,6]
y=[8,10,12]

plt.figure(figsize=(4,6),dpi=300)

plt.savefig('Tahmid.png',dpi=300)

plt.plot(x,y,'g.--',label='2x')
print(m)
plt.plot(m[:4],m[:4]**2,label='x^2')
```

Bar Charts


```python
labels=['A','B','C']
values=[1,4,2]
x=plt.bar(labels,values)

plt.figure(figsize=(6,4))
x[0].set_hatch('o')
x[2].set_hatch('/')
plt.show()
```
