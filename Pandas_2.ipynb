**MISSING** **DATA**

```python
d = {'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]}
df = pd.DataFrame(d)
df
```

```python
df.dropna()    #It drops if row have any nan values
```

```python
df.dropna(axis=1)             #It drops if column have any nan values
```

```python
df.dropna(thresh=2)     #It would keep only the rows which have atleast 2 non nan values
```

```python
df.fillna(value='Fill Value')    #It fill up the nan value
```

```python
df['A'].fillna(value=df['A'].mean())
```

**GROUPby**

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
        'Person':['Sam','Charlie','Amy','Venessa','Carl','Sarah'],
        'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
df
```

```python
bycomp = df.groupby('Company')
```

```python
bycomp.mean()
```

```python
bycomp.sum()
```

```python
bycomp.std()
```

```python
bycomp.sum().loc['GOOG']
```

```python
df.groupby('Company').sum().loc['MSFT']
```

```python
df.groupby('Company').count()
```

```python
df.groupby('Company').max()
```

```python
df.groupby('Company').min()
```

```python
df.groupby('Company').describe()
```

```python
df.groupby('Company').describe().transpose()
```

```python
df.groupby('Company').describe().transpose()['FB']
```

**Merging** **Joining** **and** **Concatenating**

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
```

```python
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                    'B':['B4','B5','B6','B7'],
                    'C':['C4','C5','C6','C7'],
                    'D':['D4','D5','D6','D7']})
```

```python
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                    'B':['B8','B9','B10','B11'],
                    'C':['C8','C9','C10','C11'],
                    'D':['D8','D9','D10','D11']})
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1,df2,df3])   #concat the rows
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
left = pd.DataFrame({'key':['K0','K1','K2','K3'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})
```

```python
right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                      'C':['C0','C1','C2','C3'],
                      'B':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how='inner',on='key')
```

```python
pd.merge(left,right,on=['key1','key2'])   #merge with two same columns
```

```python
left = pd.DataFrame({'A':['A0','A1','A2'],
              'B':['B0','B1','B2']},
               index=['K0','K1','K2'])
```

```python
right = pd.DataFrame({'C':['C0','C2','C3'],
              'D':['D0','D2','D3']},
               index=['K0','K2','K3'])
```

```python
left.join(right)
```

```python
left.join(right,how='outer')
```
